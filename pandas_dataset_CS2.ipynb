```
#1)average salary of employee by designation
df.groupby("DESIGNATION")["SALARY"].mean()
```

```
#2)count of employees by department wise
df["DESIGNATION"].value_counts()
```

```
#3)who is the most experienced person
df.sort_values("PAST EXP",ascending=False).head(1).loc[:,["FIRST NAME","LAST NAME","DESIGNATION","PAST EXP"]]
```

```
#4)top 5 rated employees
df.sort_values("RATINGS",ascending=False).head(5).loc[:,["FIRST NAME","LAST NAME","DESIGNATION","RATINGS"]]
```

```
#5)who is the one who doesnt take leave
df[df['LEAVES USED'].isnull()]
```

```
#6 top most paid employee
df.sort_values("SALARY",ascending=False).head(1).loc[:,["FIRST NAME","LAST NAME","DESIGNATION","SALARY"]]
```

```
#7)find who is the fresher and belongs to which department
df[df["PAST EXP"]==0].loc[:,["FIRST NAME","LAST NAME","DESIGNATION","PAST EXP"]]
```

```
#8)who is the youngest employee working
df.sort_values("AGE",ascending= True).head(10)
```

```
#9)find the salary > 50k and exp > 5 in it field
df[df["UNIT"]== "IT"][(df["SALARY"]> 50000) & (df["PAST EXP"] > 5)].head(5)
#df[(df["SALARY"]> 50000) & (df["PAST EXP"] > 5)]
```

```
#10)AVERAGE LEAVES REMAINING BY DESIGNATION
df.groupby("DESIGNATION")["LEAVES REMAINING"].mean()
```
